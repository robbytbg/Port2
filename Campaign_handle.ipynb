```python
import pandas as pd
```

```python
df=pd.read_excel('/content/crowdfunding.xlsx')
df
```

```python
df[['category', 'sub_category']] = df['category & sub-category'].str.split('/', expand=True)

df = df.drop(columns=['category & sub-category'])
df
```

```python
distinct_sub_categories = df['sub_category'].unique()

# Creating a DataFrame for sub-categories with unique IDs
sub_category_data = pd.DataFrame({'sub_category': distinct_sub_categories})
sub_category_data['sub_category_id'] = range(1, len(distinct_sub_categories) + 1)

# Displaying the sub-category DataFrame
print(sub_category_data)
```

```python
sub_category_data.to_csv("sub_category_data.csv", index=False)
```

```python
distinct_categories = df['category'].unique()

# Creating a DataFrame for categories with unique IDs
category_data = pd.DataFrame({'category': distinct_categories})
category_data['category_id'] = range(1, len(distinct_categories) + 1)

# Displaying the category DataFrame
print(category_data)
```

```python
category_data.to_csv("category_data.csv", index=False)
```

```python
df = pd.merge(df, category_data, on='category', how='left')

# Merge with the 'sub_category' column
df = pd.merge(df, sub_category_data, on='sub_category', how='left')

# Displaying the resulting DataFrame
df.head(5)
```

```python
df.info()
```

```python
df['launched_at'].nunique
```

```python
from datetime import datetime as dt
df["launched_at"] = pd.to_datetime(df["launched_at"],unit='s').dt.strftime('%Y-%m-%d')
df["deadline"] = pd.to_datetime(df["deadline"],unit='s').dt.strftime('%Y-%m-%d')
df.head()
```

```python
df=df.drop(['staff_pick','spotlight','category','sub_category'],axis =1)
df.head()
```

```python
df.to_csv("campaign.csv", index=False)
```
