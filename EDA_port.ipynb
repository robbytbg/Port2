```python
byContinent=df.groupby('Continent')[['2022 Population']].sum().sort_values(by="2022 Population",ascending=False)
byContinent=byContinent.reset_index()
```

```python
import numpy as np

colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")

wp = { 'linewidth' : 1, 'edgecolor' : "green" }

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:,.0f})".format(pct, absolute)

fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(byContinent['2022 Population'],
                                  autopct = lambda pct: func(pct, byContinent['2022 Population']),
                                  labels = byContinent['Continent'],
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

ax.legend(wedges, byContinent['Continent'],
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")

plt.show()
```
